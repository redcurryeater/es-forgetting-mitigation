```python
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
import matplotlib.pyplot as plt

ea = EventAccumulator("logs/beta1_iter150")
ea.Reload()
```

```python
import os
os.makedirs("figures", exist_ok=True)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

events = ea.Scalars("reward/mean")
ax1.plot([e.step for e in events], [e.value for e in events])
ax1.set_xlabel("Step")
ax1.set_ylabel("Reward Mean")
ax1.set_title("Reward Mean")

events = ea.Scalars("kl/eval")
ax2.plot([e.step for e in events], [e.value for e in events], color='orange')
ax2.set_xlabel("Step")
ax2.set_ylabel("KL Eval")
ax2.set_title("KL Eval")

plt.tight_layout()
plt.savefig("figures/reward_kl_figure.pdf")
```
